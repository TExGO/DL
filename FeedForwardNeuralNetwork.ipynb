```python
import numpy as np
import tensorflow as tf
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense,InputLayer
from keras.utils import to_categorical
```

```python
(x_train,y_train), (x_test,y_test) = mnist.load_data()

x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

x_train = x_train.reshape((x_train.shape[0],28*28))
x_test = x_test.reshape((x_test.shape[0],28*28))

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
model = Sequential()
model.add(InputLayer(shape=(28*28,)))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(10,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.2)
```

```python
test_loss, test_acc = model.evaluate(x_test,y_test)
print(f'Test Accuracy: {test_acc}')
```

```python
prediction = model.predict(x_test)
prediction_class = np.argmax(prediction,axis=1)
true_class = np.argmax(y_test,axis=1)
```

```python
import matplotlib.pyplot as plt

def plot_pre(index):
  plt.imshow(x_test[index].reshape(28,28),cmap='grey')
  plt.title(f'Predicted: {prediction_class[index]}, True: {true_class[index]}')
  plt.show()

for i in range(5):
  plot_pre(i)
```

```python
from keras.datasets import cifar10
```

```python
(x_train,y_train),(x_test,y_test) = cifar10.load_data()

x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

x_train = x_train.reshape((x_train.shape[0],32*32*3))
x_test = x_test.reshape((x_test.shape[0],32*32*3))

y_train = to_categorical(y_train,10)
y_test = to_categorical(y_test,10)
```

```python
model = Sequential()
model.add(InputLayer(shape=(32*32*3,)))
model.add(Dense(512,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.2)
```

```python
test_loss , test_acc = model.evaluate(x_test,y_test)
print(f'Test Accuracy: {test_acc}')
```

```python
prediction = model.predict(x_test)
prediction_class = np.argmax(prediction,axis=1)
true_class = np.argmax(y_test,axis=1)
```

```python
import matplotlib.pyplot as plt

def plot_pre(index):
  plt.imshow(x_test[index].reshape(32,32,3))
  plt.title(f'Predicted: {prediction_class[index]}, True:{true_class[index]}')
  plt.show()

for i in range(5):
  plot_pre(i)
```
